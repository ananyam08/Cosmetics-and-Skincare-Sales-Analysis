# EDA

```python
import matplotlib.pyplot as plt
import seaborn as sns

avg_sales = df.groupby("Product")["Amount ($)"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_sales.values, y=avg_sales.index, palette="rocket")
plt.title(" Average Sale Amount per Transaction by Product")
plt.xlabel("Average Amount ($)")
plt.ylabel("Product")
plt.tight_layout()
plt.savefig(r"C:\Users\anany\PROJECT\Cosmetics and Skincare Sales Analysis\avg sales.png")
plt.show()
```

```python
popularity = df.groupby(["Country", "Product"]).size().unstack(fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(popularity, cmap="YlGnBu", annot=True, fmt='d', linewidths=0.5)
plt.title(" Product Popularity by Country (Sales Count)")
plt.xlabel("Product")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df["Amount ($)"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Sales Amount")

plt.subplot(1, 2, 2)
sns.histplot(df["Boxes Shipped"], bins=30, kde=True, color="salmon")
plt.title("Distribution of Boxes Shipped")

plt.tight_layout()
plt.savefig(r"C:\Users\anany\PROJECT\Cosmetics and Skincare Sales Analysis\distribution .png")
plt.show()
```

```python
correlation = df["Amount ($)"].corr(df["Boxes Shipped"])
print(f"Correlation between Amount and Boxes Shipped: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Boxes Shipped", y="Amount ($)", alpha=0.6)
plt.title("Boxes Shipped vs. Amount ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Average amount and boxes per product
df.groupby("Product")[["Amount ($)", "Boxes Shipped"]].mean().sort_values("Amount ($)", ascending=False)
```

```python
grouped = df.groupby("Product")[["Amount ($)", "Boxes Shipped"]].mean().sort_values("Amount ($)", ascending=False)

grouped.plot(kind="bar", figsize=(12, 6), title="Average Sales & Boxes per Product")
plt.ylabel("Average Value")
plt.grid(True)
plt.tight_layout()
plt.savefig(r"C:\Users\anany\PROJECT\Cosmetics and Skincare Sales Analysis\avg box shipped.png")
plt.show()
```
